```python
from minicons import scorer
from nltk.tokenize import TweetTokenizer
```

```python
word_tokenizer = TweetTokenizer().tokenize
```

```python
model = "kanishka/smolm-autoreg-bpe-babylm-1e-3"
device = "cpu"

if "gpt2" in model or "pythia" in model or "SmolLM" in model:
    BOS = True
else:
    BOS = False

lm = scorer.IncrementalLMScorer(model, device=device)
```

```python
sequences = [
    "This is a sentence.",
    "This is another sentence."
]
```

```python
# if "token score" is selected

lm.token_score(
    sequences, 
    bos_token=BOS, 
    surprisal=True,
    base_two=True,
    rank=True
)
```

```python
# if the "word score" is selected

lm.word_score_tokenized(
    sequences, 
    bos_token=BOS, 
    tokenize_function=word_tokenizer,
    surprisal=True,
    base_two=True
)
```
